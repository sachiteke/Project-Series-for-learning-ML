```python
#plot 3D lines

from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
import numpy as np

ax = plt.axes(projection='3d')

zline = np.linspace(0,20,500)
yline = np.sin(zline)
xline = np.cos(zline)

ax.plot3D(xline,yline,zline,'cyan')

zplot = 20*np.random.random(500)
yplot = np.sin(zplot) + 0.1*np.random.randn(500)
xplot = np.cos(zplot) + 0.1*np.random.randn(500)

ax.scatter3D(xplot,yplot,zplot,c=zplot, cmap='Reds')
```

```python
#Counter plotting

from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


def func(x,y):
    return np.sin(np.sqrt(x**2 + y**2))

x = np.linspace(-8,8,40)
y = np.linspace(-8,8,40)

X,Y = np.meshgrid(x,y)
z = func(X,Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X,Y,z,50,cmap='binary')
```

```python
#Surface triangulation

from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

def func(x,y):
    return np.sin(np.sqrt(x**2+y**2))

theta = 2 * np.pi * np.random.random(1000)
r = 6 * np.random.random(1000)
x = np.ravel(r * np.cos(theta))
y = np.ravel(r * np.sin(theta))
z= func(x,y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter3D(x,y,z,c=z, cmap = 'viridis', linewidth = 1)
```
